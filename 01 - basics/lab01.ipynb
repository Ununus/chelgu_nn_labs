## Градиент и градиентный спуск

$$ f(x) = 5\sum_{i=0}^{n}x^2 $$

```python
def function(x):
    return 5 * (x**2).sum()
```

```python
w = torch.tensor([8., 8.], requires_grad=True)
w.requires_grad
```

```python
y = function(w)
y
```

```python
y.backward()
```

```python
w.grad
```

```python
def gradient_descent(w, alpha, func):
    y = func(w)
    y.backward()
    w.data -= alpha * w.grad
    w.grad.zero_()
```

```python
w = torch.tensor([8., 8.], requires_grad=True)
w
```

```python
for _ in range(100):
    gradient_descent(w, 0.01, function)
w
```

```python
import matplotlib.pyplot as plt
def show_contours(func, x_lims=[-10., 10.], y_lims=[-10., 10.], x_ticks=100, y_ticks=100):
    x_step = (x_lims[1] - x_lims[0]) / x_ticks
    y_step = (y_lims[1] - y_lims[0]) / y_ticks
    X, Y = np.mgrid[x_lims[0]:x_lims[1]:x_step, y_lims[0]:y_lims[1]:y_step]
    res = []
    for x_index in range(X.shape[0]):
        res.append([])
        for y_index in range(X.shape[1]):
            x_val = X[x_index, y_index]
            y_val = Y[x_index, y_index]
            func_val = func(torch.tensor([x_val, y_val])).item()
            res[-1].append(func_val)
    res = np.array(res)
    plt.figure(figsize=(7,7))
    plt.contour(X, Y, res, 100)
```

```python
show_contours(function)
```
